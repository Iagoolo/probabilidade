```python
from lightkurve import search_targetpixelfile
from lightkurve import TessTargetPixelFile
import matplotlib.pyplot as plt
import numpy as np
```

```python
pixelfile = TessTargetPixelFile("/home/cpp/Área de Trabalho/luz/stars/TIC-419012256/MAST_2025-07-07T1828/TESS/tess2018234235059-s0002-0000000419012256-0121-s/tess2018234235059-s0002-0000000419012256-0121-s_tp.fits")
pixelfile.plot(frame=1000)
plt.show()
```

```python
lc = pixelfile.to_lightcurve(aperture_mask='all')
lc.plot()
plt.show()
```

```python
lc.scatter()
plt.show()
```

```python
flat_lc = lc.flatten()
flat_lc.scatter(s=1)
flat_lc.plot()
plt.show()
```

```python
periodogram = flat_lc.to_periodogram(method='bls', period=np.arange(1, 10, 0.001))
periodogram.plot()
plt.show()
```

```python
folder_lc = flat_lc.fold(period=periodogram.period_at_max_power)
folder_lc.plot()
plt.show()
```
