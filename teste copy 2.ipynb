```python
from lightkurve import search_targetpixelfile
import matplotlib.pyplot as plt
import numpy as np
from lightkurve import TessTargetPixelFile
```

```python
pixelfile = TessTargetPixelFile('/home/cpp/Área de Trabalho/luz_p/probabilidade/stars/MAST_2025-07-10T0006/TESS/tess2018206045859-s0001-0000000261136679-0120-s/tess2018206045859-s0001-0000000261136679-0120-s_tp.fits')
pixelfile.plot()
plt.show()
```

```python
lc = pixelfile.to_lightcurve(aperture_mask='all')
lc.plot()
plt.show()
```

```python
lc_flat = lc.flatten()
lc_flat.plot()
plt.show()
```

```python
pixelFile = search_targetpixelfile('KIC 6922244', quarter=4).download()
lc_two = pixelFile.to_lightcurve(aperture_mask=pixelFile.pipeline_mask)
lc_two.plot()
plt.show()
```

```python
flat_lc = lc_two.flatten(window_length=401)
flat_lc.plot()
plt.show()
```

```python
periodogram = flat_lc.to_periodogram(method='bls', period=np.arange(3, 5, 0.001))
periodogram.plot()
plt.show()
```

```python
folder_lc = flat_lc.fold(period=periodogram.period_at_max_power)
folder_lc.plot()
plt.show()
```
